```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.kinokuniya.co.jp/f/dsd-101005021--'
res = requests.get(url)

res.text
soup = BeautifulSoup(res.text, 'html.parser')

type(soup)
book_div = soup.select_one('.list_area_wrap')

books_a_span = book_div.select('a img.mb05, span.sale_price')

len(books_a_span)
pairs = list(zip(books_a_span[0::2], books_a_span[1::2]))
book = pairs[0]

data = []

for book in pairs:
  names, prices = book
  names
  prices
  name = names.get('alt', '').strip()

  price = prices.get_text(strip = True)

  name, price
  type(price)
  price.replace("¥", "")
  price = int(price.replace("¥", "").replace(",", ""))

  type(price)
  one_data = {'name': name,'price': price}
  
  data.append(one_data)

data
!pip3 install pandas
import pandas as pd
df = pd.DataFrame(data)
df.head()
df = df.rename(columns={'name':'書籍','price':'価格'})
df.to_csv('book.csv',index=False,encoding='utf-8-sig')
```
